# The amplitude of a wave

We can multuply the output of a generator by a value between 0 and 1

```python
import math
import itertools
import numpy as np
import matplotlib.pyplot as plt

def get_sin_oscillator(freq, amp=1, phase=0, sample_rate=44100):
    phase = (phase / 360) * 2 * math.pi
    increment = (2 * math.pi * freq)/ sample_rate
    return (math.sin(phase + v) * amp for v in itertools.count(start=0, step=increment))
```

```python
osc = get_sin_oscillator(freq=1, amp= 1, phase= 0, sample_rate=512)
samples = [next(osc) for i in range(512)]

osc2 = get_sin_oscillator(freq=1, amp= 0.4, phase= 0, sample_rate=512)
samples2 = [next(osc2) for i in range(512)]

```

```python

plt.plot(samples[:], 'g', linewidth=1 , label = "1")
plt.plot(samples2[:], 'b', linewidth=1 , label = "2")
plt.xlim(-20,532)
plt.ylim(-1.25, 1.25)
plt.tick_params(axis='both', direction='out', length=6, width=2, labelcolor='black', colors='red', grid_color='gray', grid_alpha=0.5)
plt.title("Same frequency shown with 0.4 Amplitude (blue)")
plt.grid()
plt.show()
```

# Phase

The example below in desmos shows the phase angle calculation for a phase offset wave.
We are offseting the wave by 90 Degrees.

![phase wave](desmos.png)

```python
osc3 = get_sin_oscillator(freq=1, amp= 1, phase= 0, sample_rate=512)
samples3 = [next(osc3) for i in range(512)]

osc4 = get_sin_oscillator(freq=1, amp= 1, phase= 90, sample_rate=512)
samples4 = [next(osc4) for i in range(512)]
```

```python
plt.plot(samples3[:], 'g', linewidth=1 , label = "1")
plt.plot(samples4[:], 'b', linewidth=1 , label = "2")
plt.xlim(-20,532)
plt.ylim(-1.25, 1.25)
plt.tick_params(axis='both', direction='out', length=6, width=2, labelcolor='black', colors='red', grid_color='gray', grid_alpha=0.5)
plt.title("Same frequency shown with 0.4 Amplitude (blue)")
plt.grid()
plt.show()
```
