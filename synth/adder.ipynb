# Adding waves



```python
from triangleOscillator import TriangleOscillator
from waveAdder import WaveAdder
from sineOscillator import SineOscillator
from squareOscillator import SquareOscillator
import itertools
import numpy as np
import matplotlib.pyplot as plt
import math


gen = WaveAdder(
    SineOscillator(4,0,1,512),
    SquareOscillator(7,0,1,512),
    TriangleOscillator(2,0,1,512)
)

iter(gen)
samples = [next(gen) for _ in range(512)] 

```

```python
plt.figure(figsize=(12,3))
plt.plot(samples[:], 'g', linewidth=1)
plt.xlim(-20,532)
plt.ylim(-1.25, 1.25)
plt.tick_params(axis='both', direction='out', length=6, width=2, labelcolor='black', colors='red', grid_color='gray', grid_alpha=0.5)
plt.title("3 wave forms: 1 sine(4), 1 Triangle(2), 1 Square(7)")
plt.grid()
plt.show()
```
