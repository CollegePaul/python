# Ossillators

step size = (2*PI * Frequency) / samplerate

Using a generator with Lazy itteration we can just call the next sample when we need it. The get_sin_oscillator will return a generator function which can then be itterated over.

```python
import math
import itertools
import numpy as np
import matplotlib.pyplot as plt

def get_sin_oscillator(freq, sample_rate):
    increment = (2 * math.pi * freq)/ sample_rate
    return (math.sin(v) for v in itertools.count(start=0, step=increment))
```

```python
osc = get_sin_oscillator(freq=1, sample_rate=512)
samples = [next(osc) for i in range(512)]
```

```python
fig = plt.figure()
ax = plt.axes()

plt.plot(samples[:], 'g', linewidth=1)
plt.xlim(-20,532)
plt.ylim(-1.25, 1.25)
plt.tick_params(axis='both', direction='out', length=6, width=2, labelcolor='black', colors='red', grid_color='gray', grid_alpha=0.5)
plt.title("1 kh sine wave with 512 samples")
plt.grid()
plt.show()
```

```python
osc2 = get_sin_oscillator(freq=4, sample_rate=512)
samples2 = [next(osc2) for i in range(512)]
```

```python
fig = plt.figure()
ax = plt.axes()

plt.plot(samples[:], 'g', linewidth=1 , label = "1")
plt.plot(samples2[:], 'b', linewidth=1 , label = "2")
plt.xlim(-20,532)
plt.ylim(-1.25, 1.25)
plt.tick_params(axis='both', direction='out', length=6, width=2, labelcolor='black', colors='red', grid_color='gray', grid_alpha=0.5)
plt.title("1 hz and 4 hz sine wave")
plt.grid()
plt.show()
```
