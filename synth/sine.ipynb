# Using the SineOscillator class

In the example below we will initize an oscillator with a sample rate of 512 and fequency of 4, then change this without needing to make a new osscilator.

```python
from sineOscillator import SineOscillator
import itertools
import numpy as np
import matplotlib.pyplot as plt
import math

osc = SineOscillator(4,0,1,512)
osc._initialize_osc()
osc.freq = 4


samples = [next(osc) for i in range(256)]
osc.freq = 6
osc.amp = 0.8
samples2 =  [next(osc) for i in range(256)]
samples.extend(samples2)
```

```python
plt.figure(figsize=(12,3))
plt.plot(samples[:], 'g', linewidth=1)
plt.xlim(-20,532)
plt.ylim(-1.25, 1.25)
plt.tick_params(axis='both', direction='out', length=6, width=2, labelcolor='black', colors='red', grid_color='gray', grid_alpha=0.5)
plt.title("Changing frequency from 4 to 6 after 256 samples, Amplitue also changing to 0.6")
plt.grid()
plt.show()
```
