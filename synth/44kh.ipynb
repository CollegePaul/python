# 44 khz sampling

```python
import math
import itertools
import numpy as np
import matplotlib.pyplot as plt

def get_sin_oscillator(freq, sample_rate):
    increment = (2 * math.pi * freq)/ sample_rate
    return (math.sin(v) for v in itertools.count(start=0, step=increment))
```

```python
osc = get_sin_oscillator(freq=20, sample_rate=44100)
samples = [next(osc) for i in range(44100)]
```

```python


plt.figure(figsize=(12,3))
plt.plot(samples[:], 'g', linewidth=1)
plt.xlim(-200,45000)
plt.ylim(-1.25, 1.25)

plt.tick_params(axis='both', direction='out', length=6, width=2, labelcolor='black', colors='red', grid_color='gray', grid_alpha=0.5)
plt.title("20 kh sine wave with 44100 samples")
plt.grid()
plt.show()
```

```python
samples = [next(osc) for i in range(4000)]
```

```python


plt.figure(figsize=(12,3))
plt.plot(samples[:], 'g', linewidth=1)
plt.xlim(-50,4050)
plt.ylim(-1.25, 1.25)

plt.tick_params(axis='both', direction='out', length=6, width=2, labelcolor='black', colors='red', grid_color='gray', grid_alpha=0.5)
plt.title("20 kh sine showing 4000 samples at 44100 sample rate")
plt.grid()
plt.show()
```
