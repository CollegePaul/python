# Input

Sometimes you will need to get information from the user.

```python
name = input("Who are you? ")
print("hello " + name)
```

```python
age = input("how old are you?")
print(age)
```

This may not work as expected, why?

```python
nextyear = age + 1   #this is because input is always a string
print(nextyear)      #and we cant add a number and a string
```

The error shows str and int cannot be joined together.

What will fix it?

```python
nextyear = int(age) + 1   #using int() this will convert the string to a number
print(nextyear)
```

# An inportant solution

How can you make sure a person enters a number.
Use int() , however if they do enter a letter it will crash
as we cant change "hello" into a number.
To solve this we use a try - except block, this will stop it from crashing.

```python
while True:
  n1 = input("Enter a number: ")
  n2 = input("Enter 2nd number: ")
  try:
    n1 = int(n1)
    n2 = int(n2)
    break
  except:
    print("you must enter a number")

print("Thank you for entering 2 numbers")
```
